```python
import sqlite3
con = sqlite3.connect("news.db")
cur = con.cursor()

cur.execute("""
    SELECT content
    FROM table2
    ORDER BY id ASC
""")
#corpus는 db에 저장 되어있는 기사들이 저장 되어 있는 리스트
# fetchall()하면 리스트 안에 튜플 형식으로 들어있으므로 하나씩 꺼내서 리스트에 담아준다.
news = cur.fetchall()
corpus = [_[0] for _ in news]
```

```python
from collections import defaultdict
from konlpy.tag import Kkma, Hannanum

# 품사 tagset 확인
# Kkma().tagset, Hannanum.tagset

pos_tag = Kkma().pos

kkmaTDM = defaultdict(lambda:defaultdict(int))

# [_[0] for _ in pos_tag(corpus[0]) if _[1] in ["XR"]]
for d in corpus:
    for t in [_[0] for _ in pos_tag(d) 
             if _[1].startswith("N") or _[1].startswith("V")
             if len(_[0]) > 1]:
        kkmaTDM[t][d] += 1
```

```python
from math import log2
pSeed = ["노력", "한국", "평화"]
nSeed = ["규제", "일본", "비판"]

N = len(corpus)

pSeedDF = list(kkmaTDM[pSeed[0]]) # "노력" 이 나온 문서 번호 리스트
nSeedDF = list(kkmaTDM[nSeed[0]]) # "규제" 가 나온 문서 번호 리스트

pProbY = len(pSeedDF) / N # 전체 문서에서 positive 단어 나온 문서의 수
nProbY = len(nSeedDF) / N # 전체 문서에서 negative 단어 나온 문서의 수

posPMI = defaultdict(float) # positive 단어 나온것의 PMI
negPMI = defaultdict(float) # negative 단어 나온 것의 PMI

# -무한대 < PMI < min(p(x), p(y))
# -1 < NPMI < 1
# log(P(x,y) / P(x)P(y)) / - log(P(x,y))

# PMI 정규화한 것
posNPMI = defaultdict(float)
negNPMI = defaultdict(float)

for t in kkmaTDM:
    termDF = list(kkmaTDM[t]) # term t가 나온 문서 리스트 
    pX = len(termDF) / N
    posXY = (len(set(pSeedDF).intersection(termDF)) + 1) / N # smoothing 처리 -> 공집합이 나오게 되면 0이 되서 계산이 안되므로
    negXY = (len(set(nSeedDF).intersection(termDF)) + 1) / N
    
    posPMI[t] = log2(posXY / (pX * nProbY))
    negPMI[t] = log2(negXY / (pX * nProbY))
    
    posNPMI[t] = posPMI[t] / -log2(posXY)
    negNPMI[t] = negPMI[t] / -log2(negXY)
#     P(x ㅌ X) = df(t), t ㅌ V(kkmaTDM.keys) / N
#     P(y ㅌ Y) = df(pSeed[0]), y ㅌ V(kkmaTDM.keys) / N
#     P(x, y) = df(t) ^ df(pSeed[0]) / N
```

```python
sorted(negNPMI.items(), key=lambda x:x[1], reverse=True)
```

```python
SO = defaultdict(float)
nSO = defaultdict(float) # normalize한 SO

for t in kkmaTDM:
    SO[t] += posPMI[t] - negPMI[t]
    nSO[t] += posNPMI[t] - negNPMI[t]
```

```python
from math import log2
pSeed = ["노력", "한국", "평화"]
nSeed = ["규제", "일본", "비판"]

N = len(corpus)

posPMI = defaultdict(lambda:defaultdict(float)) # 단어마다 있으니까 defaultdict가 속에 들어가야....
negPMI = defaultdict(lambda:defaultdict(float)) # negative 단어 나온 것의 PMI


# PMI 정규화한 것
posNPMI = defaultdict(lambda:defaultdict(float))
negNPMI = defaultdict(lambda:defaultdict(float))

for i in range(len(pSeed)):
    pSeedDF = list(kkmaTDM[pSeed[i]]) # "노력" 이 나온 문서 번호 리스트
    nSeedDF = list(kkmaTDM[nSeed[i]]) # "규제" 가 나온 문서 번호 리스트

    pProbY = len(pSeedDF) / N # 전체 문서에서 positive 단어 나온 문서의 수
    nProbY = len(nSeedDF) / N # 전체 문서에서 negative 단어 나온 문서의 수


    for t in kkmaTDM:
        termDF = list(kkmaTDM[t]) # term t가 나온 문서 리스트
        pX = len(termDF) / N
        posXY = (len(set(pSeedDF).intersection(termDF)) + 1) / N # smoothing 처리 -> 공집합이 나오게 되면 0이 되서 계산이 안되므로
        negXY = (len(set(nSeedDF).intersection(termDF)) + 1) / N

        posPMI[pSeed[i]][t] = log2(posXY / (pX * nProbY))
        negPMI[nSeed[i]][t] = log2(negXY / (pX * nProbY))

        posNPMI[pSeed[i]][t] = posPMI[pSeed[i]][t] / -log2(posXY)
        negNPMI[nSeed[i]][t] = negPMI[nSeed[i]][t] / -log2(negXY)
```

```python
SO = defaultdict(float)
nSO = defaultdict(float) # normalize한 SO

for t in kkmaTDM:
    for i in range(len(pSeed)):
        SO[t] += posPMI[pSeed[i]][t] - negPMI[nSeed[i]][t]
        nSO[t] += posNPMI[pSeed[i]][t] - negNPMI[nSeed[i]][t]
```

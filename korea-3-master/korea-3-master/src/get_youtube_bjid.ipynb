```python
import re
```

```python
from functools import reduce
```

```python
def get_id_channel():
    pattern11 =re.compile('"shortBylineText":.+?:"WEB_PAGE_TYPE_BROWSE"')
    html = download('get','https://www.youtube.com/')
    final =pattern11.findall(html.text)
    p =[x.split(',') for x in final]
    channel_id = [[_[0].split('"text":')[1],_[2].split('"url":')[1]] for _ in p]
    pa = re.compile('"')
    se =set()
    [se.add(x) for x in [(pa.sub('',_[0]),pa.sub('',_[1])) for _ in channel_id]]
    return se
```
